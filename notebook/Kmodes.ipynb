```python
from kmodes.kmodes import KModes
import math

class KMode:
    
    def __init__(self,verbose=False,k_max_try=10):
        # configurations
        self.__verbose = 1 if verbose else 0
        self.__k_max_try = k_max_try
    
    def fit(self,data):
        cost , assignations = [] , []
        for k in range(1,self.__k_max_try+1):
            print("processing k: ",k)
            kmode = KModes(n_clusters=k, init = "random", n_init = self.__k_max_try + 1 - k, verbose=self.__verbose)
            kmode.fit_predict(data)
            cost.append( kmode.cost_ )
            assignations.append( kmode.labels_ )
        print("cost: ",cost)
```

```python
kmode = KMode()
kmode.fit(data)
```
