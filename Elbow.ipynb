```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import argrelextrema

"""
    Return the array possition where the elbow point is located
    Input: values[list]
    Output: [int]
"""
def find_optimal_k(values):
    # Compute the slope and y-intercept of the top line formula
    x1, y1 = 1, values[0]
    x2, y2 = len(values), values[-1]
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1

    # Compute the distance from each point to the top line
    distances = []
    for x, y in zip(range(1, len(values)+1), values):
        # Compute the variables for the current point
        b_data = y - m * x
        m_inverted = -1 / m
        b_intersection = y - m_inverted * x

        # Compute the point of intersection between the top line and the perpendicular line
        px = (b_intersection - b) / (m - m_inverted)
        py = m_inverted * px + b_intersection

        # Compute the distance between the point and the top line
        distance = np.sqrt((px - x)**2 + (py - y)**2)
        distances.append(distance)

    # Find the index of the maximum distance
    max_distance_index = np.argmax(distances)

    # Return the optimal k value
    return max_distance_index + 1
```

```python
# example 1
Y = [217.64,68.42,16.23,12.69,9.6,7.16,5.08,3.84,2.83]
K = find_optimal_k(Y)

print("K: ",K)

plt.plot(range(1,len(Y)+1),Y)
```
