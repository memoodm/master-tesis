```python
import matplotlib.pyplot as plt

import pandas as pd

from app.Categoricer import Categoricer
from app.MCA import MCA
from app.KMeans import Kmeans
from app.KModes import Kmodes
```

```python
data = pd.read_csv(r"./data/data.csv")
data = data.iloc[:, :6]
data
```

```python
# generate categories
categoricer_model = Categoricer(show_graph_solution=False)
categories = categoricer_model.generate_cagegories(data)

# apply MCA
mca_model = MCA()
coordinates_row , _ = mca_model.getCoordinates(data)

# apply kmeans to mca points
kmeans_model = Kmeans()
K,_ = kmeans_model.get_K_from_points(coordinates_row)

# apply kmodes
kmodes_model = Kmodes()
labels = kmodes_model.segmentate_for_K(categories,K)
```
